# Regression Demo

## Generating Data

```python
from machine_learning_jh.data.regression import Line
line_generator = Line(n_points=200, noise_std_ratio=0.2)
x, y = line_generator.generate_quadratic()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.show()
```

## Linear Regression

```python
from machine_learning_jh.models.regression import MultivariateRegressor

linear_regressor = MultivariateRegressor()
linear_regressor.fit(x, y)
y_pred_linear = linear_regressor.predict(x)
```

```python
from machine_learning_jh.metrics.regression import mse

linear_mean_squared_error = mse(y, y_pred_linear)
print(f'Mean squared error of linear regressor: {linear_mean_squared_error:.3f}')
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred_linear, 'r-')
plt.show()
```

## Polynomial Regression

```python
from machine_learning_jh.models.regression import PolynomialRegressor

polynomial_regressor = PolynomialRegressor(rank=2)
polynomial_regressor.fit(x, y)
y_pred_quadratic = polynomial_regressor.predict(x)
```

```python
quadratic_mean_squared_error = mse(y, y_pred_quadratic)
print(f'Mean squared error of linear regressor:    {linear_mean_squared_error:.3f}')
print(f'Mean squared error of quadratic regressor: {quadratic_mean_squared_error:.3f}')
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred_quadratic, 'r-')
plt.show()
```

## Periodic Data

```python
x_periodic, y_periodic = line_generator.generate_periodic()
```

```python
polynomial_regressor_periodic = PolynomialRegressor(rank=6)
polynomial_regressor_periodic.fit(x_periodic, y_periodic)
y_pred_periodic = polynomial_regressor_periodic.predict(x_periodic)

periodic_mean_squared_error = mse(y_periodic, y_pred_periodic)
print(f'Mean squared error: {periodic_mean_squared_error:.2f}')
```

```python
plt.scatter(x_periodic, y_periodic)
plt.plot(x_periodic, y_pred_periodic, 'r-')
plt.show()
```
